```python
df = pd.read_csv("netflix_titles.csv")
df.head()  # mostra as 5 primeiras linhas
```

```python
print("Total de títulos na base:", df.shape[0])
```

```python
df['type'].value_counts().plot(kind='bar', color='tomato')
plt.title("Distribuição de Filmes e Séries")
plt.ylabel("Quantidade")
plt.show()
```

```python
df['country'].value_counts().head(5).plot(kind='barh', color='skyblue')
plt.title("Top 5 Países com Mais Títulos")
plt.xlabel("Quantidade")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,4))
plt.title("Lançamentos por Ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.show()
```
